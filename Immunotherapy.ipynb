```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import model_from_json
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
from sklearn import preprocessing
```

```python
df = pd.read_excel(r"Immunotherapy.xlsx", sheet_name='ImmunoDataset')
```

```python
df.head()
```

```python
df.info()
```

```python
## Removing Duplicates
dup = df[df.duplicated()]
df.drop_duplicates()
dup.shape
```

```python
df['Result_of_Treatment'] = df['Result_of_Treatment'].astype('category')
```

```python
# checking null values
df.isnull().sum()
```

```python
## Now creating feature vector and target variable
x = df.drop(['Result_of_Treatment'], axis=1)
y = df['Result_of_Treatment']
```

```python
# input dimention
x_dim = x.shape[1]
y_dim = y.shape[0]
```

```python
# scaling
x_s = preprocessing.scale(x)
y_s = preprocessing.scale(y)
```

```python
le = preprocessing.LabelEncoder()
y_s = le.fit_transform(y_s)
```

```python
# Train test plitting
x_train, x_test, y_train, y_test = train_test_split(x_s, y_s, test_size=0.25, random_state=4)
```

# Creating a Deep learning sequential model

```python
model = Sequential()
```

```python
model.add(Dense(12, input_dim=x_dim, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the model
model.fit(x_train, y_train, epochs=200, batch_size=10)
```

```python
# scores
score = model.evaluate(x_train, y_train)
```

```python
print("\n%s: %.2f%%" % (model.metrics_names[1], score[1]*100))
```

```python
model.save('Immo.h5')
```
